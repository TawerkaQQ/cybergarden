## Calling `imAnonymizer` object on an image

```python
import cv2
import numpy as np
from anonympy.images import imAnonymizer
from google.colab.patches import cv2_imshow

img = cv2.imread('Data/3.jpeg')
anonym = imAnonymizer(img)

blurred = anonym.face_blur(kernel = (31, 31), shape = 'r', box = 'r')
pixaled = anonym.face_pixel(shape = 'c', box = None)
SaP = anonym.face_SaP(shape = 'c', box = 'c') # Salt And Pepper Noise 

all = np.hstack([blurred, pixaled, SaP])

cv2_imshow(all)
```
